```python
path = r'.\custom_images'
filenames = os.listdir(path)
imgdata = []
for file in filenames:
    temp = Image.open(os.path.join(path,file))
    temp = np.array(temp)
    # print(temp.shape)
    imgdata = imgdata + [temp]
imgdata = np.array(imgdata)
```

```python
imgdata.shape
```

```python
imgdata = (imgdata.reshape(imgdata.shape[0],imgdata.shape[1]*imgdata.shape[2]*imgdata.shape[3]))
```

```python
imgdata.shape
```

```python
# test = pd.DataFrame(imagedata)
# tuple(imgdata.shape)
imgdata[0]
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(imgdata, square, test_size=0.3, random_state=123)
```

```python
# type(X_train)
X_train.shape
```

### 

```python
model1 = DecisionTreeClassifier()
# model2 = RandomForestClassifier(random_state=123,n_estimators=60)
model1.fit(X_train,y_train)
# model2.fit(X_train,y_train)
```

```python
score_filter1 = model1.score(X_test, y_test)
score_filter1
```

```python
predictions = model1.predict(X_test)
predictions[:10]
```

```python
pre = pd.DataFrame()
pre['prediction'] = predictions
pre['real'] = y_test
pre.head(100)
# 
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy:%s'% accuracy_score(y_test,predictions))
```
