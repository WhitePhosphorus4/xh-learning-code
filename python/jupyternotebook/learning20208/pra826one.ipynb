```python
# 导入鸢尾花数据集
from sklearn.datasets import load_iris
iris = load_iris()
print(type(iris))
# display(iris)

print('iris数据集特征')
print(iris.data[:10])

print('iris数据集标签')
print(iris.target[:3])
```

```python
# 导入模型包，加载模型
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
```

```python
# 数据集
x_train,y_train = iris.data[:120],iris.target[:120]
# 验证集
x_dev,y_dev = iris.data[120:],iris.target[120:]

print(x_train[:5])
print(y_train[:5])
```

```python
# 模型训练
clf.fit(x_train,y_train)
```

```python
# 模型测试的测试结果
predictions = clf.predict(x_dev)
predictions[:10]
```

```python
# 准确评价指标
from sklearn.metrics import accuracy_score
print('Accuracy:%s'% accuracy_score(y_dev,predictions))
```

```python
from matplotlib import pyplot as plt
plt.plot(x_dev, y_dev,color='blue')
plt.twinx()
# 预测值，predict()预测函数
plt.plot(x_dev, predictions,color='red')
plt.show()
```
