# 航空客户流失模型
将航空公司用户划分为三类：未流失、准流失、已流失

使用关联性划分特征

```python
import pandas as pd
import numpy as np
#读数据
input_file = 'air_data_done1.csv'
data = pd.read_csv(input_file)

# 数据转换
# data['入会时间'] = pd.to_datetime(data['入会时间'])
data["LOAD_TIME"] = pd.to_datetime(data["LOAD_TIME"])
# data["FFP_DATE"] = pd.to_datetime(data["FFP_DATE"])
data["FIRST_FLIGHT_DATE"] = pd.to_datetime(data["FIRST_FLIGHT_DATE"])
# data["LAST_FLIGHT_DATE"] = pd.to_datetime(data["LAST_FLIGHT_DATE"])
data = data.replace({'GENDER':{'男':1,'女':2}})

#选取特征
data['单位里程票价'] = (data['SUM_YR_1'] + data['SUM_YR_2'])/data['SEG_KM_SUM']
data['单位里程积分'] = (data['P1Y_BP_SUM'] + data['L1Y_BP_SUM'])/data['SEG_KM_SUM']
data['开始坐飞机时长'] = data["LOAD_TIME"] - data["FIRST_FLIGHT_DATE"]
# data['最后一次坐飞机间隔'] = data["LOAD_TIME"] - data["LAST_FLIGHT_DATE"]
#筛选出老客户(飞行次数大于6次的为老客户)
data = data[data['FLIGHT_COUNT'] > 6]
#选择特征
# data = data[['FFP_TIER','飞行次数比例','AVG_INTERVAL',
#              'avg_discount','EXCHANGE_COUNT','Eli_Add_Point_Sum','单位里程票价','单位里程积分']]
data.shape
```

```python
# 简单查看异常，进行初步筛查,可以确定要去除EP_SUM_YR_1
data.describe().T
```

```python
data.info()
```

```python
# 打标签
data['飞行次数比例'] = data['SUM_YR_2'] / data['SUM_YR_1'] #第二年飞行次数与第一年飞行次数的比例
col = np.zeros((data.shape[0]))

for i,val in enumerate(data['飞行次数比例']):
    if val < 0.5:
        col[i] = 0
    if val >= 0.5 and val <0.9:
        col[i] = 1
    if val >= 0.9:
        col[i] = 2

data['客户类型'] = col
# 用完就丢
data_delete = ['SUM_YR_2','SUM_YR_1','飞行次数比例']
data = data.drop(data_delete,axis=1)
# data.head()
display(col[:100])
```

```python
data = data.drop('EP_SUM_YR_1',axis=1)
```

## 特征关联性分析

```python
# 分析特征的相关性
import matplotlib.pyplot as plt
import seaborn as sns
# 解决中文乱码
sns.set_style('whitegrid',{'font.sans-serif':['simhei','AriLAST_FLIGHT_DATEal']}) 
corrdf = data.corr()
display(corrdf)
plt.figure(figsize=(12, 12))
# 热力图
sns.heatmap(data.corr())
plt.show()
```

```python
asdf = corrdf['客户类型'].sort_values(ascending=False)
asdf
```

## 数据处理
需要出去特征相关性分析中关联不是特别大的特征


#### 特征处理

```python
# 去除相关系数小于0.1的特征
for i,key in enumerate(asdf.keys()):
    if abs(asdf[i])<0.1 :
        data = data.drop(key,axis=1,inplace=False)
#         print(key)

# 去除x与x之间相关性大于0.9，且与y的相关性的绝对值比较小的特征
delete_columns = ['FFP_DATE','WORK_PROVINCE','LOAD_TIME','WORK_COUNTRY',
                  'WORK_CITY','FIRST_FLIGHT_DATE','LAST_FLIGHT_DATE','入会时间',
                  '开始坐飞机时长']
    
data = data.drop(delete_columns,axis=1,inplace= False)
data.head()
```

```python
# 变量划分
y = data['客户类型']
X = data.drop('客户类型',axis=1)
# display(X.head())
# display(y.head())
```

```python
# 标准化
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)

X_done = scaler.transform(X)
cacl = pd.DataFrame(X_done,columns=data.drop('客户类型',axis=1).keys()).head()
cacl.columns
```

```python
# 模型训练
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_done, y, test_size=0.3, random_state=123)

# clf = RandomForestClassifier(random_state=123)
model1 = DecisionTreeClassifier()
model2 = RandomForestClassifier(random_state=123,n_estimators=60)
model1.fit(X_train,y_train)
model2.fit(X_train,y_train)
# score_filter2 = model2.score(X_test,y_test)
# print(score_filter2)
```

```python
# 调参
from sklearn import metrics
scores = []
for i in range(1, 100):
    rd = RandomForestClassifier( n_estimators=i,random_state=123)
    rd.fit(X_train,y_train)
    scores.append(metrics.accuracy_score(y_test, rd.predict(X_test)))
plt.figure(figsize=(10, 8))
plt.plot(range(1,100), scores, marker='o')
plt.xlabel('Number of n_estimators')
plt.ylabel('score')
plt.show() 
```

### 模型评估

```python
# 自带评分
score_filter1 = model1.score(X_test, y_test)
score_filter2 = model2.score(X_test,y_test)
print(score_filter1)
print(score_filter2)
```

```python
# 交叉验证
from sklearn.model_selection import cross_val_score

print('交叉验证')
print("-----------------------------")
print("决策树")
accuracy=cross_val_score(model1,X_test,y_test,scoring='accuracy',cv=5)
print('准确率：{:.2f}%'.format(accuracy.mean()*100))
precision=cross_val_score(model1,X_test,y_test,scoring='precision_weighted',cv=5)
print('精确度：{:.2f}%'.format(precision.mean()*100))
recall=cross_val_score(model1,X_test,y_test,scoring='recall_weighted',cv=5)
print('召回率：{:.2f}%'.format(recall.mean()*100))
f1=cross_val_score(model1,X_test,y_test,scoring='f1_weighted',cv=5)
print('F1  值：{:.2f}%'.format(f1.mean()*100))

print("-----------------------------")
print("随机森林")
accuracy=cross_val_score(model2,X_test,y_test,scoring='accuracy',cv=5)
print('准确率：{:.2f}%'.format(accuracy.mean()*100))
precision=cross_val_score(model2,X_test,y_test,scoring='precision_weighted',cv=5)
print('精确度：{:.2f}%'.format(precision.mean()*100))
recall=cross_val_score(model2,X_test,y_test,scoring='recall_weighted',cv=5)
print('召回率：{:.2f}%'.format(recall.mean()*100))
f1=cross_val_score(model2,X_test,y_test,scoring='f1_weighted',cv=5)
print('F1  值：{:.2f}%'.format(f1.mean()*100))
```
